**Five Key Findings about Crime in Chicago**  
  
1) While there have been changes in crime rates over 2017 to 2018, it is clear that the order of crime reports by type is relatively stable with theft, battery and criminal damage being the most reported crimes.   
  
2) The neighborhoods with the highest reports of crime include the downtown area as well as the west side of Chicago. 
  
3) Violent crimes such as homicides and batteries tend to be reported more in impoverished neighborhoods by zipcodes, specifically those with higher percentage of the population with no internet access, higher percentage of the population in extreme poverty (with income below 50% of the poverty line) and a lower median income.

4) These neighborhoods with higher report of violent crimes by zipcodes also tend to have less white population.

5) Non-violent crimes are on average reported in neighbodhoods with more white population, less of the population with no internet access and in extreme poverty.

Taken together, this seems to suggest that crime reports occur in areas where there is more people as well as in areas that are less resourced. The data seems to show that the crimes that occur (or at least reported) in these two areas are somewhat different based on the community characteristics. Hence in order to deal with crime in the city, there needs to be an effort to identify and understand the different clusters of crime types in the different community areas and how they relate to the community's characteristics. Perhaps the areas where there is higher traffic gives more opportunity for crime, but crime occur in areas that are less resourced due to greater incentive for gang affiliation. These data cannot give us conclusive answers to how we can deal with crime in the city, but illuminates some patterns and trends that can be further analyzed.

**Caveats**

Firstly, this entirely analysis is based on reported crime and there is no information about crimes that occur but are not reported. It may be the case that certain crimes are not being reported or crimes that are misreported and worse if these errors in data disporportionally affect some types of crime (e.g. sex offenses) or some neighborhoods more than others.

Secondly, this analysis is merely correlational and descriptive with no causal inference.
  
Finally, the merging of the data may not be entirely accurate. I merged the crime and ACS data on zipcode information using the latest ACS-5 year estimate 2012-2017 but does not cover 2018. Hence this analysis assumes that the characteristics of the community has not changed from 2017 to 2018 which may not be true. In addition, it also assumes accuracy of the spatial join of the zipcode shape files with the coordinates provided in the crime data.
